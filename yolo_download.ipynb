```python
### GPT's  ###
import torch
from torch.utils.data import DataLoader
from torch.utils.tensorboard import SummaryWriter
from tqdm import tqdm
from yolov5.models import YOLOv5
from yolov5.utils.datasets import LoadImagesAndLabels
from yolov5.utils import google_utils
from yolov5.utils.torch_utils import select_device

def train_yolov5(train_data, batch_size, num_epochs, data_yaml, cfg_file, weights, results_name):
    # Tensorboard 
    writer = SummaryWriter()

    #  
    device = select_device()
    hyp = {'lr0': 0.01, 'lrf': 0.01, 'momentum': 0.937, 'weight_decay': 0.0005, 'warmup_epochs': 3.0, 'warmup_momentum': 0.8, 'warmup_bias_lr': 0.1,
           'box': 0.05, 'cls': 0.5, 'cls_pw': 1.0, 'obj': 1.0, 'obj_pw': 1.0, 'iou_t': 0.2, 'anchor_t': 4.0, 'fl_gamma': 0.0, 'hsv_h': 0.015, 'hsv_s': 0.7, 'hsv_v': 0.4,
           'degrees': 0.0, 'translate': 0.1, 'scale': 0.5, 'shear': 0.0, 'perspective': 0.0, 'flipud': 0.0, 'fliplr': 0.5, 'mosaic': 1.0, 'mixup': 0.0, 'copy_paste': 0.0}

    #  
    model = YOLOv5(cfg_file, imgsz=416)

    #    
    google_utils.attempt_download(weights)
    model.load_state_dict(torch.load(weights, map_location=device)['model'])

    #   
    dataset = LoadImagesAndLabels(train_data, imgsz=416, batch_size=batch_size, hyp=hyp, augment=True, cache_images=True, rect=True)
    dataloader = DataLoader(dataset, batch_size=batch_size, num_workers=8, pin_memory=True, collate_fn=dataset.collate_fn)

    #  
    optimizer = torch.optim.SGD(model.parameters(), lr=hyp['lr0'], momentum=hyp['momentum'], weight_decay=hyp['weight_decay'])

    #  
    global_step = 0
    for epoch in range(num_epochs):
        model.train()
        for batch_idx, (imgs, targets, _, _) in enumerate(tqdm(dataloader)):
            #   
            imgs = imgs.to(device)
            targets = targets.to(device)

            # Forward Pass
            loss, _ = model(imgs, targets)

            # Backward Pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            #    Tensorboard 
            writer.add_scalar('Loss/train', loss.item(), global_step)
            global_step += 1

    #   
    torch.save(model.state_dict(), f'{results_name}.pt')
    writer.close()

#      
train_data = '/jupyter_workspace/yolo3_test/dataset/train.txt'
batch_size = 16
num_epochs = 1
data_yaml = '/jupyter_workspace/yolo3_test/data.yaml'
cfg_file = './yolov5/models/yolov5s.yaml'
weights = 'yolov5s.pt'
results_name = 'gun_yolov5s_results'

# YOLOv5  
train_yolov5(train_data, batch_size, num_epochs, data_yaml, cfg_file, weights, results_name)
```
